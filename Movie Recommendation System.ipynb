```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
 movies.head()
```

```python
credits.head()
```

```python
movies= movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
movies= movies[['id','title','overview','genres','keywords','cast_x','crew_y']]
```

```python
movies.head()
```

```python
 movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L   
```

```python
import ast
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
 movies.head()
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def convert3(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count+=1
        else:
            break
                
    return L   
```

```python
movies['cast_x']= movies['cast_x'].apply(convert3)
```

```python
movies.head(1)
```

```python
def director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew_y']= movies['crew_y'].apply(director)
```

```python
movies.head(2)
```

```python
movies['overview']= movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
movies['genres']= movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast_x']= movies['cast_x'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew_y']= movies['crew_y'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(3)
```

```python
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast_x'] + movies['crew_y']
```

```python
movies.head(1)
```

```python
new_df= movies[['id', 'title', 'tags']]
```

```python
new_df
```

```python
new_df['tags']= new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python

from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000, stop_words='english')
```

```python
 vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)   
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
